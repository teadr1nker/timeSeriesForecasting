```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mape
def MDA(X, Y):
    if len(X) != len(Y):
        return None
    s = .0
    for i in range(1, len(X)):
        if np.sign(X[i] - X[i-1]) == np.sign(Y[i] - Y[i-1]):
            s += 1
        else:
            s -= 1
    return (1 - (s / len(X))) / 2
def metrics(res):
    benches = []
    days = [5, 10, 15, 20, 40, 252]
    for l in days:
        bench = {}
        bench['days'] = l
        bench['mse'] = round(mse(res[1][:l], res[0][:l]), 3)
        bench['mda'] = round(MDA(res[1][:l], res[0][:l]), 3)
        bench['mse x mda'] = round(bench['mse'] * bench['mda'], 3)
        
        benches.append(bench)
    return benches
def MDA(X, Y):
    if len(X) != len(Y):
        return None
    s = .0
    for i in range(1, len(X)):
        if np.sign(X[i] - X[i-1]) == np.sign(Y[i] - Y[i-1]):
            s += 1
        else:
            s -= 1
    return (1 - (s / len(X))) / 2
def metrics(res):
    benches = []
    days = [5, 10, 15, 20, 40, 252]
    for l in days:
        bench = {}
        bench['days'] = l
        bench['mse'] = round(mse(res[1][:l], res[0][:l]), 3)
        bench['mda'] = round(MDA(res[0][:l], res[1][:l]), 3)
        bench['mape'] = round(mape(res[1][:l], res[0][:l]), 3)
        
        benches.append(bench)

    return benches
        
```
