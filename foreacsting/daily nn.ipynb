```python
import yfinance as yf
from CNN_test import test_CNN
from RNN_test import testRNN
import matplotlib.pyplot as plt
from tree_test import testTree
from forest_test import testForest
import pandas as pd
from sklearn.metrics import root_mean_squared_error
import numpy as np
```

```python
def MDA(X, Y):
    if len(X) != len(Y):
        return None
    s = .0
    for i in range(1, len(X)):
        if np.sign(X[i] - X[i-1]) == np.sign(Y[i] - Y[i-1]):
            s += 1
        else:
            s -= 1
    return (1 - (s / len(X))) / 2
```

```python
data = yf.download('gld', '2020-01-01', '2022-01-01')
```

```python
benchmarks = []
res1 = test_CNN(data)
res2 = testRNN(data)
plt.plot(res1[0][0], label='CNN')
plt.plot(res2[0][0], label='simple RNN')
plt.plot(res2[1][0], label='LSTM')
plt.plot(res2[2][0], label='GRU')
plt.plot(res1[0][1], label='Real', ls='--', c='r')

benchmarks.append({'name': "CNN",
                   'mse': root_mean_squared_error(res1[0][0], res1[0][1]),
                   'mda': MDA(res1[0][0], res1[0][1])})    
benchmarks.append({'name': "simpleRNN",
                   'mse': root_mean_squared_error(res2[0][0], res2[0][1]),
                   'mda': MDA(res1[0][0], res1[0][1])})
benchmarks.append({'name': "LSTM",
                   'mse': root_mean_squared_error(res2[1][0], res2[1][1]),
                   'mda': MDA(res1[0][0], res1[0][1])})
benchmarks.append({'name': "GRU",
                   'mse': root_mean_squared_error(res2[2][0], res2[2][1]),
                   'mda': MDA(res1[0][0], res1[0][1])})
    
plt.legend(loc='best')
```

```python
resultTree = testTree(data)
resultForest = testForest(data)
plt.plot(resultForest[0][0], label="Forest")
plt.plot(resultTree[0][0], label="Tree")
plt.plot(resultTree[0][1], label="Real")

plt.legend(loc='best')
plt.savefig('tree+forest.png')
    
benchmarks.append({"name": 'Tree',
                       "mse": root_mean_squared_error(resultTree[0][0], resultTree[0][1]),
                       "mda": MDA(resultTree[0][0], resultTree[0][1])})

benchmarks.append({"name": 'Forest',
                       "mse": root_mean_squared_error(resultForest[0][0], resultForest[0][1]),
                       "mda": MDA(resultTree[0][0], resultTree[0][1])})
```
